```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('heights.csv')
df.head()
```

```python
max_t=df.height.quantile(0.95)
```

```python
min_t=df.height.quantile(0.05)
```

```python
df.shape
```

```python
df=df[(df['height']>min_t) & (df['height']<max_t)]
```

```python
df.shape
```

#Removing outliers from complex dataset

```python
df=pd.read_csv('bhp.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
min_t,max_t=df.price_per_sqft.quantile([0.001,0.999])
min_t,max_t
```

```python
df[df.price_per_sqft<min_t]
```

```python
df[df.price_per_sqft>max_t]
```

```python
df=df[(df.price_per_sqft>min_t) & (df.price_per_sqft<max_t)]
```
