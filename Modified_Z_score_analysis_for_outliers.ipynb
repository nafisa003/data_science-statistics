```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('movie_revenues.csv')
df.head()
```

```python
df.revenue.describe()
```

```python
df['revenue_mln']=df['revenue'].apply(lambda x: x/10**6)
df.head()
```

```python
_,mean,std,*_=df.revenue_mln.describe()
```

```python
mean,std
```

```python
def get_z_score(val,mean,std):
  return (val-mean)/std
```

```python
df['zscores']=df.revenue_mln.apply(lambda x: get_z_score(x,mean,std))

```

```python
df.head()
```

```python
df[df.zscores>3]
```

#Modified Z Score

```python
def get_mean_abs_dev(s):
  med=np.median(s)
  diff=abs(s-med)
  MAD=np.median(diff)
  return MAD
```

```python
MAD=get_mean_abs_dev(df.revenue_mln)
```

```python
median=np.median(df.revenue_mln)
```

```python
MAD,median
```

```python
df['modified_zscore']=df.revenue_mln.apply(lambda x: ((0.6745 * (x-median))/MAD))
```

```python
df.head()
```

```python
df[df.modified_zscore>3.5]
```
